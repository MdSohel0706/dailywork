**READING XML FILE AND CREATING A DATA FROM THE DATA FOUND FROM THE FILE**

```python
upload = files.upload()
```

```python
import xml.etree.ElementTree as et
xtree = et.parse("cd_catalog.xml")
```

```python
def transform_xml(xml_doc):
    xroot = xml_doc.getroot()
    rows = []
    for cd in xroot:
        title = cd.find("TITLE").text 
        artist = cd.find("ARTIST").text
        country = cd.find("COUNTRY").text
        company = cd.find("COMPANY").text
        price = cd.find("PRICE").text
        year = cd.find("YEAR").text
        rows.append({"TITLE" : title, "ARTIST" : artist, "COUNTRY" : country, 
                   "COMPANY" : company, "PRICE" : price, "YEAR" : year}) 
    df = pd.DataFrame(rows)
    return df
```

```python
out_df = transform_xml(xtree)
out_df.head()
```

```python
def find_parameters(xml_doc):
    xroot = xml_doc.getroot()
    parameters = []
    for tags in xroot[0]:
        parameters.append(tags.tag)
    return parameters
```

```python
columns = find_parameters(xtree)
print(columns)
```
